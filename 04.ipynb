```python
# import thư viện
from PIL import Image
from imgtools import *
```

```python
# Đường Dẫn Ảnh
my_path = 'C:/Users/Dell/Documents/srv_cv/img/01/01.jpg'
```

```python
# Đọc Ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100, 50, 170, 200)
# Cắt Vùng Ảnh
cropped_image = img.crop(region)
```

```python
display(cropped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (150, 150)
img.paste(cropped_image, paste_position)
display(img)
```
