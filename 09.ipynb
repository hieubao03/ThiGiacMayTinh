```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Duong Danh Anh
my_path = 'C:/Users/Dell/Documents/srv_cv/img/01/01.jpg'
# Doc Anh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyển sang ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=150)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# lam noi bat duong bien
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```
