```python
# Import lớp Image
from PIL import Image
from imgtools import *
from IPython.display import display # to display images
```

```python
# Đường Dẫn Ảnh
my_path = 'C:/Users/Dell/Documents/srv_cv/img/01/01.jpg'
```

```python
# Đọc ảnh 
img = load_image(my_path)

# Resize
# img = img.resize((250,250))
```

```python
display(img)
```

```python
# Chuyển từ ảnh RGB sang ảnh xảm
gray_image =  img.convert('L')
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
merged_image = Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```

```python
# VD Đảo lại vị trí màu red_band thì sẽ trông như bên dưới
merged_image = Image.merge("RGB", ( green_band, blue_band, red_band))
display(merged_image)
```
